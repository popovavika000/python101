```python
import pandas as pd

# Загрузка данных
file_path = 'dst-3.0_16_1_hh_database.csv'  
data = pd.read_csv(file_path, sep=';')  # Чтение данных с разделителем ;

# Функция для преобразования уровня образования
def extract_education_level(education_string):
    # Извлечение первые два слова из строки
    words = education_string.split()[:2]
    
    # Сопоставление полученные слова с категориями
    if 'Высшее' in words:
        return 'высшее'
    elif 'Неоконченное' in words:
        return 'неоконченное высшее'
    elif 'Среднее' in words and 'специальное' in words:
        return 'среднее специальное'
    elif 'Среднее' in words:
        return 'среднее'
    else:
        return 'неизвестно'  # Для строк, которые не подходят под категории

# Применение функции к столбцу "Образование и ВУЗ" и создание нового столбец "Образование"
data['Образование'] = data['Образование и ВУЗ'].apply(extract_education_level)

# Проверяем уникальные значения в новом столбце
print(data['Образование'].unique())

# Удаляем столбец "Образование и ВУЗ"
data.drop(columns=['Образование и ВУЗ'], inplace=True)

# Печать первых 5 строк, чтобы убедиться в результатах
print(data.head())
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
import pandas as pd
import re

# Загрузка данных
file_path = 'dst-3.0_16_1_hh_database.csv'  
data = pd.read_csv(file_path, sep=';')  # Чтение данных с разделителем ;

# Функция для преобразования пола
def extract_gender(gender_string):
    if 'Мужчина' in gender_string:
        return 'М'
    elif 'Женщина' in gender_string:
        return 'Ж'
    return 'Неизвестно'

# Функция для преобразования возраста
def extract_age(age_string):
    # Извлекаем число, представляющее возраст
    age_match = re.search(r'\d+', age_string)
    if age_match:
        return int(age_match.group(0))  # Возвращаем возраст как целое число
    return None

# Применяем функцию для создания нового признака "Пол"
data['Пол'] = data['Пол, возраст'].apply(lambda x: extract_gender(x.split(' , ')[0]))

# Применяем функцию для создания нового признака "Возраст"
data['Возраст'] = data['Пол, возраст'].apply(lambda x: extract_age(x.split(' , ')[1]))

# Удаляем столбец "Пол, возраст"
data.drop(columns=['Пол, возраст'], inplace=True)

# Печать первых 5 строк, чтобы убедиться в результатах
print(data.head())

# Проверка уникальных значений в новом столбце "Пол" и "Возраст"
print(data['Пол'].unique()) 
print(data['Возраст'].unique()) 
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
import pandas as pd
import numpy as np
import re

# Загрузка данных
file_path = 'dst-3.0_16_1_hh_database.csv' 
data = pd.read_csv(file_path, sep=';')  # Чтение данных с разделителем ;

# Функция для преобразования опыта работы в месяцы
def extract_experience(experience_string):
    if pd.isna(experience_string) or experience_string == 'Не указано':
        return np.nan  # Возвращаем NaN для пропусков и строк "Не указано"
    
    # Извлекаем опыт работы (годы и месяцы)
    years = 0
    months = 0
    
    # Применяем регулярные выражения для поиска лет и месяцев
    year_match = re.search(r'(\d+)\s*года*', experience_string)
    if year_match:
        years = int(year_match.group(1))  # Извлекаем число лет
        
    month_match = re.search(r'(\d+)\s*месяцев*', experience_string)
    if month_match:
        months = int(month_match.group(1))  # Извлекаем число месяцев
    
    # Переводим все в месяцы
    total_months = years * 12 + months
    return total_months

# Применяем функцию для создания нового признака "Опыт работы (месяц)"
data['Опыт работы (месяц)'] = data['Опыт работы'].apply(extract_experience)

# Удаляем старый столбец "Опыт работы"
data.drop(columns=['Опыт работы'], inplace=True)

# Печать первых 5 строк для проверки
print(data.head())

# Проверка на уникальные значения в новом столбце "Опыт работы (месяц)"
print(data['Опыт работы (месяц)'].unique())

```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

```python
import pandas as pd
import numpy as np
import re

# Список городов-миллионников
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 
                  'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']

# Функция для извлечения города
def extract_city(location_string):
    if pd.isna(location_string):
        return 'другие'  # если нет данных, считаем город - "другие"
    # Извлекаем город до первой запятой
    city = location_string.split(',')[0].strip()
    
    # Проверяем, является ли город миллионником
    if city in million_cities:
        return 'город-миллионник'
    elif city == 'Москва' or city == 'Санкт-Петербург':
        return city  # Москва или Спб
    else:
        return 'другие'  # остальные города

# Функция для извлечения готовности к переезду
def extract_relocation_readiness(location_string):
    if pd.isna(location_string):
        return False  # если нет данных, считаем, что не готов к переезду
    # Проверяем фразы о готовности к переезду
    if re.search(r'готов.*переезду|хочу.*переехать', location_string, re.IGNORECASE):
        return True
    elif re.search(r'не готов.*переезду', location_string, re.IGNORECASE):
        return False
    return False  # по умолчанию считаем, что не готов к переезду

# Функция для извлечения готовности к командировкам
def extract_travel_readiness(location_string):
    if pd.isna(location_string):
        return False  # если нет данных, считаем, что не готов к командировкам
    # Проверяем фразы о готовности к командировкам
    if re.search(r'готов.*командировкам|готова.*командировкам', location_string, re.IGNORECASE):
        return True
    elif re.search(r'не готов.*командировкам', location_string, re.IGNORECASE):
        return False
    return False  # по умолчанию считаем, что не готов к командировкам

# Применение функций к столбцу "Город, переезд, командировки"
data['Город'] = data['Город, переезд, командировки'].apply(extract_city)
data['Готовность к переезду'] = data['Город, переезд, командировки'].apply(extract_relocation_readiness)
data['Готовность к командировкам'] = data['Город, переезд, командировки'].apply(extract_travel_readiness)

# Удаление старого столбца "Город, переезд, командировки"
data.drop(columns=['Город, переезд, командировки'], inplace=True)

# Проверка результатов
print(data[['Город', 'Готовность к переезду', 'Готовность к командировкам']].head())
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
import pandas as pd

# Чтение данных из файла CSV
data = pd.read_csv('dst-3.0_16_1_hh_database.csv', sep=';', on_bad_lines='skip')

# Выводим названия колонок, чтобы найти, как называются нужные столбцы
print(data.columns)

# Предположим, что в данных есть столбцы "Занятость" и "График" — адаптируем под них
# Замените 'Занятость' и 'График' на актуальные имена столбцов из вашего датасета
# Например, если столбцы называются 'employment' и 'schedule', замените их в коде.

# Пример данных для столбцов "Занятость" и "График"
data['Занятость'] = data['Занятость'].fillna('')  # Заполняем пустые значения, если есть
data['График'] = data['График'].fillna('')

# Списки категорий для занятости и графика
employment_categories = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
schedule_categories = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']

# Функция для преобразования признака "Занятость"
def one_hot_encode_employment(employment_string):
    result = {}
    for category in employment_categories:
        result[category] = category in employment_string
    return result

# Функция для преобразования признака "График"
def one_hot_encode_schedule(schedule_string):
    result = {}
    for category in schedule_categories:
        result[category] = category in schedule_string
    return result

# Применяем OHE для столбцов "Занятость" и "График"
employment_encoded = data['Занятость'].apply(one_hot_encode_employment)
schedule_encoded = data['График'].apply(one_hot_encode_schedule)

# Преобразуем результат в DataFrame и объединяем с исходным DataFrame
employment_df = pd.DataFrame(employment_encoded.tolist())
schedule_df = pd.DataFrame(schedule_encoded.tolist())

# Объединяем DataFrame с исходными данными
data = pd.concat([data, employment_df, schedule_df], axis=1)

# Удаляем оригинальные столбцы "Занятость" и "График"
data.drop(columns=['Занятость', 'График'], inplace=True)

# Проверка результата
print(data.head())
```

6. (2 балла) Наконец, мы добрались до самого главного и самого важного - признака заработной платы **"ЗП"**. 
В чем наша беда? В том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:
* 30000 руб.
* 50000 грн.

Нам бы хотелось видеть заработную плату в единой валюте, например, в рублях. Возникает вопрос, а где взять курс валют по отношению к рублю?

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
import pandas as pd

currency_data = pd.read_csv('dst-3.0_16_1_hh_database.csv', sep=';', on_bad_lines='skip')

# Посмотрим на столбцы и первые строки данных
print(currency_data.head())
print(currency_data.columns)

# Преобразуем столбец "Обновление резюме" в datetime (это наша дата)
currency_data['Обновление резюме'] = pd.to_datetime(currency_data['Обновление резюме'], format='%d.%m.%Y %H:%M')

# Пример данных с резюме
resume_data = {
    'ЗП': ['30000 руб.', '50000 грн.', '550 USD', '40000 руб.', '1000 KZT'],
    'Обновление резюме': ['16.04.2019 15:59', '12.04.2019 08:42', '16.04.2019 08:35', '08.04.2019 14:23', '22.04.2019 10:32']
}

# Создаем DataFrame с резюме
resume_df = pd.DataFrame(resume_data)

# Преобразуем столбец "Обновление резюме" в datetime
resume_df['Обновление резюме'] = pd.to_datetime(resume_df['Обновление резюме'], format='%d.%m.%Y %H:%M')

# Функция для извлечения суммы и валюты из строки зарплаты
def extract_salary_and_currency(salary_str):
    salary = ''.join([c if c.isdigit() else '' for c in salary_str])  # Извлекаем цифры
    currency = ''.join([c if c.isalpha() else '' for c in salary_str]).upper()  # Извлекаем валюту и делаем верхний регистр
    return int(salary), currency

# Применяем функцию для создания новых столбцов с зарплатой и валютой
resume_df[['ЗП_сумма', 'Валюта']] = resume_df['ЗП'].apply(lambda x: pd.Series(extract_salary_and_currency(x)))

# Извлекаем только дату из столбца "Обновление резюме"
resume_df['Дата_обновления'] = resume_df['Обновление резюме'].dt.date

# Преобразуем столбец 'Дата_обновления' в datetime для объединения с курсами
resume_df['Дата_обновления'] = pd.to_datetime(resume_df['Дата_обновления'])

# Объединим данные с курсами валют (по дате и валюте)
merged_df = pd.merge(resume_df, currency_data, how='left', left_on=['Дата_обновления', 'Валюта'], right_on=['Обновление резюме', 'currency'])

# Если курс не найден, заменяем его на 1 для рубля
merged_df['rate'] = merged_df['rate'].fillna(1)

# Переводим зарплату в рубли
merged_df['ЗП (руб)'] = merged_df['ЗП_сумма'] * merged_df['rate']

# Удаляем ненужные столбцы
merged_df.drop(columns=['ЗП', 'Обновление резюме', 'ЗП_сумма', 'Валюта', 'currency', 'rate'], inplace=True)

# Выводим результат
print(merged_df[['ЗП (руб)', 'Дата_обновления']])


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re  # Импортируем модуль для регулярных выражений

# Чтение данных из файла CSV
data = pd.read_csv('dst-3.0_16_1_hh_database.csv', sep=';', on_bad_lines='skip')

# Выводим названия колонок, чтобы найти нужный столбец
print(data.columns)

# Функция для преобразования опыта работы в месяцы
def extract_experience_months(experience):
    # Пропускаем NaN и значения с типом float
    if isinstance(experience, str):
        years = 0
        months = 0
        
        # Ищем количество лет
        years_match = re.search(r'(\d+)\s*год', experience)
        if years_match:
            years = int(years_match.group(1)) * 12  # переводим в месяцы

        # Ищем количество месяцев
        months_match = re.search(r'(\d+)\s*месяц', experience)
        if months_match:
            months = int(months_match.group(1))
        
        return years + months
    return np.nan  # Если не строка, возвращаем NaN

# Применяем функцию для извлечения месяцев
data['Опыт работы (месяц)'] = data['Опыт работы'].apply(extract_experience_months)

# Строим гистограмму и коробчатую диаграмму для анализа
plt.figure(figsize=(12, 6))

# Гистограмма
plt.subplot(1, 2, 1)
sns.histplot(data['Опыт работы (месяц)'], bins=30, kde=True)
plt.title('Гистограмма: Опыт работы (в месяцах)')
plt.xlabel('Опыт работы (месяцы)')
plt.ylabel('Частота')

# Коробчатая диаграмма
plt.subplot(1, 2, 2)
sns.boxplot(x=data['Опыт работы (месяц)'])
plt.title('Коробчатая диаграмма: Опыт работы (в месяцах)')
plt.xlabel('Опыт работы (месяцы)')

plt.tight_layout()
plt.show()

# Описание статистики
mode_experience = data['Опыт работы (месяц)'].mode()[0]
min_experience = data['Опыт работы (месяц)'].min()
max_experience = data['Опыт работы (месяц)'].max()
median_experience = data['Опыт работы (месяц)'].median()

# Выводим статистику
print(f"Мода опыта работы (месяцы): {mode_experience}")
print(f"Минимальный опыт работы (месяцы): {min_experience}")
print(f"Максимальный опыт работы (месяцы): {max_experience}")
print(f"Медианный опыт работы (месяцы): {median_experience}")

# Интервал, в котором находится опыт работы для большинства соискателей
iqr = data['Опыт работы (месяц)'].quantile(0.75) - data['Опыт работы (месяц)'].quantile(0.25)
lower_bound = data['Опыт работы (месяц)'].quantile(0.25) - 1.5 * iqr
upper_bound = data['Опыт работы (месяц)'].quantile(0.75) + 1.5 * iqr

print(f"Интервал опыта работы для большинства соискателей: {lower_bound:.0f} - {upper_bound:.0f} месяцев")

# Анализ аномалий
outliers = data[data['Опыт работы (месяц)'] > upper_bound]
print(f"Число аномальных значений (выше верхней границы): {len(outliers)}")
print(outliers[['Опыт работы', 'Опыт работы (месяц)']])

    
   


```

7. Постройте сводную таблицу, иллюстрирующую зависимость **медианной** желаемой заработной платы от возраста (**"Возраст"**) и образования (**"Образование"**). На полученной сводной таблице постройте **тепловую карту**. Проанализируйте тепловую карту, сравнив показатели внутри групп.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dst-3.0_16_1_hh_database.csv', sep=';', on_bad_lines='skip')

# Очистка данных
df.dropna(subset=['ЗП', 'Возраст', 'Образование'], inplace=True)  # Удаляем строки с пустыми значениями
df['ЗП'] = df['ЗП'].replace({' руб.': '', ' бел.руб.': '', ',': '', 'KZT': '', 'USD': '', '€': ''}, regex=True)

# Преобразуем столбец 'ЗП' в числовой формат
df['ЗП'] = pd.to_numeric(df['ЗП'], errors='coerce')

# Преобразуем возраст в числовой формат (если данные в возрасте представлены как строки с возрастом, убираем ненужные символы)
df['Возраст'] = df['Возраст'].replace({' лет': '', 'год': '', 'лет': ''}, regex=True).astype(float)

# Фильтруем данные по желаемой зарплате меньше 1 миллиона рублей
df_filtered = df[df['ЗП'] < 1000000]

# Создаем сводную таблицу с медианными значениями ЗП по возрасту и образованию
pivot_table = df_filtered.pivot_table(values='ЗП', index='Возраст', columns='Образование', aggfunc='median')

# Построение тепловой карты
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='coolwarm', cbar_kws={'label': 'Медианная зарплата (руб.)'})

# Настроим внешний вид
plt.title('Тепловая карта медианной заработной платы в зависимости от возраста и образования')
plt.xlabel('Образование')
plt.ylabel('Возраст')

plt.tight_layout()
plt.show()
```

6. В результате анализа мы обнаружили потенциальные выбросы в признаке **"Возраст"**. Это оказались резюме людей чересчур преклонного возраста для поиска работы. Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore


df = pd.read_csv('dst-3.0_16_1_hh_database.csv', sep=';', on_bad_lines='skip')

# Преобразуем столбец "Возраст" в числовой формат (если он не числовой)
df['Возраст'] = pd.to_numeric(df['Возраст'], errors='coerce')

# Удаляем строки с пропущенным возрастом
df = df.dropna(subset=['Возраст'])

# Логарифмируем возраст
log_age = np.log(df['Возраст'])

# Строим гистограмму возраста в логарифмическом масштабе
plt.figure(figsize=(10, 6))
sns.histplot(log_age, kde=True, color='skyblue')

# Добавляем линии для среднего и границ интервала трех сигм (с учетом 4 сигм для правой границы)
mean_age = log_age.mean()
std_age = log_age.std()

plt.axvline(mean_age, color='k', lw=2, label='Среднее')
plt.axvline(mean_age - 4*std_age, color='r', linestyle='--', label='Левая граница (4 сигмы)')
plt.axvline(mean_age + 4*std_age, color='r', linestyle='--', label='Правая граница (4 сигмы)')

# Добавляем подписи и легенду
plt.title('Распределение возраста в логарифмическом масштабе')
plt.xlabel('Логарифм возраста')
plt.ylabel('Частота')
plt.legend()
plt.show()

# Оценка асимметрии
# Если распределение правосторонне асимметричное, то хвост будет тянуться вправо.
# Это видно по тому, что правая граница (сигма) значительно смещена в сторону высоких значений.
print("Средний возраст:", mean_age)
print("Стандартное отклонение:", std_age)

# Вычисление z-отклонений для логарифмированного возраста
z_scores = zscore(log_age)

# Отбор выбросов (выбросы — это значения с z-отклонением больше 4 или меньше -4)
outliers = df[np.abs(z_scores) > 4]

# Печать таблицы с выбросами
display(outliers)

# Удаление выбросов из данных
df_cleaned = df[np.abs(z_scores) <= 4]

print(f"Удалено {len(outliers)} строк с выбросами.")

```
